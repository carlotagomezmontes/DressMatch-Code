# Create First Item Function

```python
def createfirstitem():
  moreitems = input("\nDo you want to insert a new item?('yes' or 'no'): ")
  while moreitems == "yes":
    ask_the_user_for_type_of_item = input("What type of clothes is it? ('bottom', 'top', 'all in one' or 'shoes' ): ").lower()
    if ask_the_user_for_type_of_item != "shoes":
      name, brand, length, color, season = createnew()
    else:
      name, brand, color, season = createnewshoes()

    if  ask_the_user_for_type_of_item == "bottom":
      newbottom = Bottom(name, brand, length, color, season)
      bottom_list.append(newbottom) 
      all_items.append(newbottom)
    elif ask_the_user_for_type_of_item == "top":
      newtop = Top(name, brand, length, color, season)
      tops_list.append(newtop)
      all_items.append(newtop) 
    elif ask_the_user_for_type_of_item == "all in one":
      newallinone = AllInOne(name, brand, length, color, season)
      allinone_list.append(newallinone)
      all_items.append(newallinone) 
    elif ask_the_user_for_type_of_item == "shoes":
      newshoes = Shoes(name, brand, color, season)
      shoes_list.append(newshoes)
      all_items.append(newshoes)
    
    moreitems = input("\nDo you want to insert another new item?('yes' or 'no'): ")

  print("\nYou are done with the inserting items part. Now you are ready to continue with the next part")        
```

# Get Outfit Function

```python
def getoutfit(bottom_list, tops_list, allinone_list, shoes_list):
  ask_the_user_first_condition = input("Do you want a 'set' or do you prefere an 'all in one'? ").lower()
  list_bottoms_available = []
  list_tops_available = []
  list_shoes_available = []
  list_allinone_available = []

  # set
  if ask_the_user_first_condition == "set":
    ask_the_user_second_condition= input("Do you want to choose an outfit based on 'color' or 'season'? ").lower()

    # filter by color
    if ask_the_user_second_condition == "color":

      color_type = input("Choose a color: 'white', 'black', 'blue', 'green', ...: ").lower()

      for bottom in bottom_list:
        if bottom.color == color_type:
          list_bottoms_available.append(bottom)
      
      for top in tops_list:
        if top.color == color_type:
          list_tops_available.append(top)
      
      for shoes in shoes_list:
        if shoes.color == color_type:
          list_shoes_available.append(shoes)
       
    # filter by season
    if ask_the_user_second_condition == "season":
      chosen_season = input("Choose a season: 'winter' or 'summer': ").lower()

      for bottom in bottom_list:
        if bottom.season == chosen_season:
          list_bottoms_available.append(bottom)
      
      for top in tops_list:
        if top.season == chosen_season:
          list_tops_available.append(top)
      
      for shoes in shoes_list:
        if shoes.season == chosen_season:
          list_shoes_available.append(shoes)

  # all in one
  if ask_the_user_first_condition == "all in one":
    ask_the_user_second_condition = input("Do you want to choose an outfit based on 'color' or 'season'? ").lower()

    # filter by color
    if ask_the_user_second_condition == "color":

      color_type = input("Choose a color: 'white', 'black', 'blue', 'green', ...: ").lower()
      
      for allinone in allinone_list:
        if allinone.color == color_type:
          list_allinone_available.append(allinone)
        
      for shoes in shoes_list:
        if shoes.color == color_type:
          list_shoes_available.append(shoes)
      
      if list_shoes_available == []:
        for shoes in shoes_list:
          if shoes.color == "white":
            list_shoes_available.append(shoes)
  

    # filter by season
    if ask_the_user_second_condition == "season":
    
      chosen_season = input("Choose a season: 'winter' or 'summer': ").lower()
      for allinone in allinone_list:
        if allinone.season == chosen_season:
          list_allinone_available.append(allinone)
        
      for shoes in shoes_list:
        if shoes.season == chosen_season:
          list_shoes_available.append(allinone)

  return ask_the_user_first_condition, list_bottoms_available, list_tops_available, list_shoes_available, list_allinone_available
```
